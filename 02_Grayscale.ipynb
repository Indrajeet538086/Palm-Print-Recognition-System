# Grayscale 

Applying grayscaling over ROI to remove the lightning factor from the image.

```python
import cv2
import os
from pathlib import Path
```

```python

COLOR_ROI_DIR = Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\01_Symmetric_Palm_ROI_3")
FINAL_GRAY_DIR = Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\02_Grayscale 2 latest work")

os.makedirs(FINAL_GRAY_DIR, exist_ok=True)

if __name__ == "__main__":
    # Scan the processed Color ROI folder for all generated images
    processed_color_images = list(COLOR_ROI_DIR.rglob('*.png'))
    
    if not processed_color_images:
        print(f"⚠️ No color images found in {COLOR_ROI_DIR}. Run Step 1 first!")
    else:
        print(f"🔄 Starting separate Grayscale transformation for {len(processed_color_images)} items...")
        converted_count = 0
        
        for color_img_path in processed_color_images:
            # Read the cropped color ROI image
            color_roi = cv2.imread(str(color_img_path))
            if color_roi is None: continue
            
            # PURE CONVERSION: Strip down channels into 1-channel Grayscale Matrix
            gray_roi = cv2.cvtColor(color_roi, cv2.COLOR_BGR2GRAY)
            
            # Maintain exact folder trees (User_ID / Handedness)
            relative_structure = color_img_path.relative_to(COLOR_ROI_DIR)
            save_path = FINAL_GRAY_DIR / relative_structure
            
            # Ensure folder layers exist
            os.makedirs(save_path.parent, exist_ok=True)
            
            # Save the clean grayscale image file
            cv2.imwrite(str(save_path), gray_roi)
            converted_count += 1
            
        print("=" * 55)
        print(f"✅ Stage 2 Complete. Clean Grayscale dataset ready.")
        print(f"Total separate grayscaled outputs saved: {converted_count}")
        print(f"Final Save Path: {FINAL_GRAY_DIR}")
        print("=" * 55)
```
