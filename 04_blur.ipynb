# Gaussian Blurring

Applying Gaussian blurring with filter size 3 to the CLAHE images to remove the extra noise from the images

```python
import cv2
import os
from pathlib import Path
```

```python
TUNING_CONFIG = {
    
    "input_clahe_dir": Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\03_Enhanced_CLAHE 5 latest work(7.0)"),
    "output_blur_dir": Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\04_Noise_Reduced 7 latest work_7-3"),
    
    
    "median_blur_kernel": 3             
}
```

```python



# Ensure final destination folder exists
os.makedirs(TUNING_CONFIG["output_blur_dir"], exist_ok=True)

if __name__ == "__main__":
    # Scan the CLAHE folder for all images recursively
    clahe_images = list(TUNING_CONFIG["input_clahe_dir"].rglob('*.png'))
    
    if not clahe_images:
        print(f"No images found in {TUNING_CONFIG['input_clahe_dir']}. Ensure Step 3 ran successfully!")
    else:
        print(f"Starting separate Median Blur filter for {len(clahe_images)} items...")
        processed_count = 0
        
        for clahe_img_path in clahe_images:
            # Read image directly as a 1-channel grayscale matrix
            clahe_roi = cv2.imread(str(clahe_img_path), cv2.IMREAD_GRAYSCALE)
            if clahe_roi is None: 
                continue
            
            # PURE BLURRING: Apply median filter to smoothly drop noise without destroying line edges
            blurred_roi = cv2.medianBlur(clahe_roi, TUNING_CONFIG["median_blur_kernel"])
            
            # Reconstruct exact identical sub-folder trees (User_ID / Handedness)
            relative_structure = clahe_img_path.relative_to(TUNING_CONFIG["input_clahe_dir"])
            save_path = TUNING_CONFIG["output_blur_dir"] / relative_structure
            
            # Ensure sub-folders exist
            os.makedirs(save_path.parent, exist_ok=True)
            
            # Save the smoothed file
            cv2.imwrite(str(save_path), blurred_roi)
            processed_count += 1
            
        print("=" * 55)
        print(f"Stage 4 Complete. Noise reduction filtering finished.")
        print(f"Total blurred outputs saved: {processed_count}")
        print(f"Final Save Path: {TUNING_CONFIG['output_blur_dir']}")
        print("=" * 55)
```
