# CLAHE

Applying CLAHE (Contrast Limited Adaptive Histogram Equalization) with clip limit 7 over grayscale images to clearly extract the principle lines from the palm region.

```python
import cv2
import os
from pathlib import Path
```

```python
TUNING_CONFIG = {
    
    "input_gray_dir": Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\02_Grayscale 2 latest work"),
    "output_clahe_dir": Path(r"C:\Users\singh\Users\singh\Desktop\Palm detection\ROI\03_Enhanced_CLAHE 5 latest work(7.0)"),
    
    # --- CLAHE Hyper-parameters ---
    "clahe_clip_limit": 7.0,            # Higher = higher contrast / line amplification
    "clahe_grid_size": (5, 5)           # Grid resolution for localized equalization
}
```

```python

# Ensure final destination folder exists
os.makedirs(TUNING_CONFIG["output_clahe_dir"], exist_ok=True)

if __name__ == "__main__":
    # Scan the grayscale folder for all images recursively
    gray_images = list(TUNING_CONFIG["input_gray_dir"].rglob('*.png'))
    
    if not gray_images:
        print(f"No images found in {TUNING_CONFIG['input_gray_dir']}. Ensure Step 2 ran successfully!")
    else:
        print(f"Starting separate CLAHE enhancement for {len(gray_images)} items...")
        processed_count = 0
        
        for gray_img_path in gray_images:
            # Read image explicitly as Grayscale (1-channel)
            gray_roi = cv2.imread(str(gray_img_path), cv2.IMREAD_GRAYSCALE)
            if gray_roi is None: 
                continue
            
            # PURE CLAHE: Create the processor object and apply it directly to the matrix
            clahe_processor = cv2.createCLAHE(
                clipLimit=TUNING_CONFIG["clahe_clip_limit"], 
                tileGridSize=TUNING_CONFIG["clahe_grid_size"]
            )
            enhanced_roi = clahe_processor.apply(gray_roi)
            
            # Reconstruct exact identical sub-folder trees (User_ID / Handedness)
            relative_structure = gray_img_path.relative_to(TUNING_CONFIG["input_gray_dir"])
            save_path = TUNING_CONFIG["output_clahe_dir"] / relative_structure
            
            # Ensure sub-folders exist
            os.makedirs(save_path.parent, exist_ok=True)
            
            # Save the contrast-enhanced file
            cv2.imwrite(str(save_path), enhanced_roi)
            processed_count += 1
            
        print("=" * 55)
        print(f"Stage 3 Complete. Contrast enhancement finished.")
        print(f"Total contrast-equalized outputs saved: {processed_count}")
        print(f"Final Save Path: {TUNING_CONFIG['output_clahe_dir']}")
        print("=" * 55)
```
